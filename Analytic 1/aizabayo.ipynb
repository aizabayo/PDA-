```python
"""Annotating special event in sales"""
def determine_seasonal_sales(monthly_sales):
    max_sales = max(monthly_sales)  # Find the maximum sales
    min_sales = min(monthly_sales)  # Find the minimum sales
    max_sales_index = monthly_sales.index(max_sales)
    min_sales_index = monthly_sales.index(min_sales)

    max_sales_month = months[max_sales_index]
    min_sales_month = months[min_sales_index]

    return {
        'Month with max sales': max_sales_month,
        'Month with min sales': min_sales_month,
        'Maximum sales': max_sales,
        'Minimum sales': min_sales
    }

result = determine_seasonal_sales(sales)

max_sales_month = result['Month with max sales']
min_sales_month = result['Month with min sales']
max_sales_value = result['Maximum sales']
min_sales_value = result['Minimum sales']

print(f'{max_sales_month} is the month has high sales equaL to: {max_sales_month} with sales of {max_sales_value}')
print(f'{min_sales_month} is the month that has very low sales equal to: {min_sales_value}')

period = [min_sales_month, max_sales_month]
Sales = [min_sales_value, max_sales_value]
plt.plot(months, sales, color='darkgreen')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.title('Highest and Lowest Sales Months')
plt.show()
```

```python
#Ranking Top 5 station based on Average crime rate above entire period
Column_df = Data_df.columns[3:-1]

Average_crime = Data_df.groupby('Station')[Column_df].mean()

# Calculate Average crime rate for each station
Average_crime['Average Crime Rate'] = Average_crime.mean(axis=1)

# Sort the data to find the top 5 stations with the highest and lowest average crime rates
High_crime_stations = Average_crime.sort_values(by='Average Crime Rate', ascending=False).head(5)
Low_crime_stations = Average_crime.sort_values(by='Average Crime Rate').head(5)

# Simplify the output DataFrames
High_crime_stations = High_crime_stations[['Average Crime Rate']].reset_index()
High_crime_stations.columns = ['Station', 'Average Crime Rate (Highest)']

Low_crime_stations = Low_crime_stations[['Average Crime Rate']].reset_index()
Low_crime_stations.columns = ['Station', 'Average Crime Rate (Lowest)']

# Print the results

print('Ranking Top 5 station based on Average crime rate')
display(High_crime_stations)
print('Top 5 stations with the highest crime rates over the entire period:')
display(High_crime_stations)

print('\nTop 5 stations with the lowest crime rates over the entire period:')
display(Low_crime_stations)
```

```python
"""Analyzing the overall crime trends in each province"""

df3 = pd.read_csv('SouthAfricaCrimeStats_v2.csv')      # Loading dataset
# Find the five most frequent provinces
top_provinces = df3['Province'].value_counts().nlargest().index

# Define custom colors for each province
colors = ['b', 'g', 'r', 'c', 'brown']  # You can change these colors as desired

# Filter the data for the top provinces
filter_data = df3[df3['Province'].isin(top_provinces)]

# Group the data by province and sum the crime counts over the years
province_crime = filter_data.groupby('Province').sum().iloc[:, 2:]

plt.figure(figsize = (10,6))
for i, province in enumerate(top_provinces):
    plt.plot(province_crime.columns, province_crime.loc[province], label=province, color=colors[i])

plt.xlabel('Years')
plt.ylabel('Total Crime Count')
plt.title('Aggregate Crime Trends for the Top 5 Provinces')
plt.legend(title='Province')
plt.xticks(rotation = 60)
plt.grid()

plt.show()
```
